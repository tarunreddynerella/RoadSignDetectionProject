```python
# Include all packages
import cv2
import numpy as np
import pandas as pd
```

```python
annotationsDataFrame = pd.read_csv("./DataAnalysis.csv")
annotationsDataFrame1 = pd.read_csv("./DataAnalysis1.csv")
annotationsDataFrame = pd.concat([annotationsDataFrame,annotationsDataFrame1], ignore_index=True)
annotationsDataFrame["width"] = annotationsDataFrame['x2'] - annotationsDataFrame['x1']
annotationsDataFrame["height"] = annotationsDataFrame['y2'] - annotationsDataFrame['y1']
annotationsDataFrame["area"] = annotationsDataFrame['width'] * annotationsDataFrame['height']
annotationsDataFrame["perimeter"] = (annotationsDataFrame['width'] + annotationsDataFrame['height'])*2
annotationsDataFrame.head()
```

```python
annotationsDataFrame.describe()
```

```python
widthData = annotationsDataFrame['width'].to_numpy()
widthData.sort()
print(np.percentile(widthData, 90))
print(np.percentile(widthData, 95))
heightData = annotationsDataFrame['height'].to_numpy()
heightData.sort()
print(np.percentile(heightData, 90))
print(np.percentile(heightData, 95))
```
